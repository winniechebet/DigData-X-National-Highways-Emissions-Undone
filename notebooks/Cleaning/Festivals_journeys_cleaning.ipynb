# Step 1: Importing, Inspecting, and Cleaning Data
Ensuring the datasets are free from missing or messy data that could interfere with processing.

### Import, inspect and clean **1st dataset**: **'Festival_journeys'**

```python
# Read 'Festival_journeys' dataset into a dataframe
journeys_df = pd.read_csv('raw_datasets/Festival_journeys_dataset.csv')
```

```python
# Inspect components of 'journeys df'
journeys_df.info()
```

```python
# Display first 10 rows of 'journeys_df'
journeys_df.head(10)
```

```python
# Drop repeated index field
journeys_df.drop(columns=['Unnamed: 0'], inplace=True)
# Verify if specified column is dropped
journeys_df.head(3)
```

```python
# Create a DataFrame with a subset of entities with non-null 'Carpooling' observations
carpooling_df = journeys_df[journeys_df['Carpooling'].notnull()].copy()
```

```python
# Inspect the 'carpooling_df'
carpooling_df.info()
carpooling_df.head()
```

```python
# Drop empty cells in 'carpooling_df'
carpooling_df.dropna(inplace=True)
# Write 'carpooling_df' into a.csv file
carpooling_df.to_csv('cleaned_datasets/Carpooling_dataset.csv', index=False)
```

```python
# Drop 'Carpooling' column from 'journeys_df'
journeys_df.drop(columns=['Carpooling'], inplace=True)
# Drop null cells in 'journeys_df'
journeys_df.dropna(inplace=True)
# Verify its dropped
journeys_df.info()
journeys_df.head()
```

```python
# Write 'journeys_df' into a .csv file
journeys_df.to_csv('cleaned_datasets/Festivals_journeys_dataset.csv', index=False)
```
