# Research Question 1

**Question:** Which modes of transport warrant the least time on the road for each unique "Origin" location? From the result of this inquiry, is there a mode of transport that proves consistent in terms of user prefrence throughout each unique origin location? Or are they varied?

**Approach:** Engineer a "Speed" feature from "Time" and "Distance" to conduct this comparison in an absolute manner


```python
# Import relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the main dataset "Festival's Journey" into a DataFrame
journeys_df = pd.read_csv("../cleaned_datasets/Festivals_journeys_dataset.csv")
journeys_df.head()
```

```python
# Verify creation of "Speed" feature
journeys_df.head()
```

#### Subset data into peak and non-peak
I want to compare which modes of transport take the shortest or longest time to reach the festival destination from different origin locations. However, these modes fall into two categories: those that traveled during peak hours and those that traveled during non-peak hours. If I make this comparison without separating the two, the insights may be inaccurate. This is because travel time includes periods when vehicles were stuck in traffic (for those that traveled during peak hours), making it easy to misinterpret the data and assume they were actively moving the entire time.

```python
journeys_df["Peak/NoPeak"].value_counts()
```

```python
# Create "Peak" and "No Peak" subsets from the "journeys_df" dataframe
peak_subset = journeys_df[journeys_df["Peak/NoPeak"] == "Peak"]
no_peak_subset = journeys_df[journeys_df["Peak/NoPeak"] == "No Peak"]
peak_subset.head()
```

```python
no_peak_subset.head()
```

```python
# Distinct Modes of Transport in the dataset
MoTs = list(journeys_df["Mode of Transport"].unique())
MoTs
```

```python
# Define function to compare travel time based on the mode of transport and origin location
def travel_time_comparison(df):
    # Inspect which Mode of Transport is the fastest regardless of the place of origin
    time_comparison = {}
    # Loop over the list of "MoTs"
    for MoT in MoTs:
        # Create a subset of "journeys_df" containing only a single, unique MoT
        unique_MoT = df[df["Mode of Transport"] == MoT]
        # Get the average "Travel Time" of the unique MoT from each unique origin. 
        time_by_MoT_and_origin = unique_MoT.groupby("Origin")["Travel Time"].mean()
        # Append "time_by_MoT_and_origin" to "unique_MoT_subsets" dictionary
        time_comparison[f"{MoT} Travel Time"] = time_by_MoT_and_origin    
        
    return time_comparison
```

```python
# Store the return value of the function as a DataFrame for peak hour data 
peak_df = pd.DataFrame(travel_time_comparison(peak_subset))
# Store the return value of the same function as a DataFrame for non-peak hour data  
no_peak_df = pd.DataFrame(travel_time_comparison(no_peak_subset))
```

```python
peak_df
```

```python
# Instantiate a list object to store the sorted values of "Travel Time" from the different origins
fastet_MoT_peak = []
# Loop over the indexes of the peak_df
for idx in peak_df.index:
    # Retrieve values of unique "Origin" and sort starting from the "MoT" with the least "Travel Time"
    sorted_times = peak_df.loc[idx].sort_values()
    fastet_MoT_peak.append(sorted_times)    
```

```python
# Create a grid of subplots
fig, axs = plt.subplots(3, 2, figsize=(10, 10))

# Set main title for the entire figure
fig.suptitle("Comparison of Travel Time by Different Modes of Transport - Peak Hours", fontsize=14, fontweight="bold")


fastet_MoT_peak[0].plot(kind="barh", 
                        title=peak_df.index[0],
                        xlabel = "Time(Hrs)",
                        color="#82964C",
                        edgecolor="black",
                        ax=axs[0, 0])

fastet_MoT_peak[1].plot(kind="barh", 
                        title=peak_df.index[1],
                        xlabel = "Time(Hrs)",
                        color="#82964C",
                        edgecolor="black",
                        ax=axs[0, 1])

fastet_MoT_peak[2].plot(kind="barh", 
                        title=peak_df.index[2],
                        xlabel = "Time(Hrs)",
                        color="#82964C",
                        edgecolor="black",
                        ax=axs[1, 0])

fastet_MoT_peak[3].plot(kind="barh", 
                        title=peak_df.index[3],
                        xlabel = "Time(Hrs)",
                        color="#82964C",
                        edgecolor="black",
                        ax=axs[1, 1])

fastet_MoT_peak[4].plot(kind="barh", 
                        title=peak_df.index[4],
                        xlabel = "Time(Hrs)",
                        color="#82964C",
                        edgecolor="black",
                        ax=axs[2, 0])

fastet_MoT_peak[5].plot(kind="barh", 
                        title=peak_df.index[5],
                        xlabel = "Time(Hrs)",
                        color="#82964C",
                        edgecolor="black",
                        ax=axs[2, 1])


plt.tight_layout()
```

Now that we've identified which modes of transport take the least time on the road from different origins, we can analyze people's transport preferences. My goal is to determine which origin locations need the most awareness regarding travel choices based on time spent on the road.

```python
# Distinct origin locations are in the dataset
origins_list = list(journeys_df["Origin"].unique()) 
origins_list
```

```python
# Loop over the list of "location" labels
mot_prefrence_groups = []
for origin in origins_list:
    # Subset "journeys_df" into a set containing only one "origin" location
    origin_mask = journeys_df[journeys_df["Origin"] == origin]
    # Group the different MoTs based on the origin
    origin_groups = origin_mask.groupby("Origin")["Mode of Transport"].value_counts()
    # Visually represent the MoT preference by people from different origin locations
    mot_prefrence_groups.append(origin_groups)
```

```python
# Create a grid of subplots
fig, axs = plt.subplots(3, 2, figsize=(10, 10))

# Set main title for the entire figure
fig.suptitle("Preferred Modes of Transport by Travelers from Different Origins", fontsize=14, fontweight="bold")


mot_prefrence_groups[0].plot(kind="barh", 
                        title=list(mot_prefrence_groups[0].index[0])[0],
                        color="#E2C43F",
                        edgecolor="black",
                        ax=axs[0, 0])

mot_prefrence_groups[1].plot(kind="barh", 
                        title=list(mot_prefrence_groups[1].index[0])[0],
                        color="#E2C43F",
                        edgecolor="black",
                        ax=axs[0, 1])

mot_prefrence_groups[2].plot(kind="barh", 
                        title=list(mot_prefrence_groups[2].index[0])[0],
                        color="#E2C43F",
                        edgecolor="black",
                        ax=axs[1, 0])

mot_prefrence_groups[3].plot(kind="barh", 
                        title=list(mot_prefrence_groups[3].index[0])[0],
                        color="#E2C43F",
                        edgecolor="black",
                        ax=axs[1, 1])

mot_prefrence_groups[4].plot(kind="barh", 
                        title=list(mot_prefrence_groups[4].index[0])[0],
                        color="#E2C43F",
                        edgecolor="black",
                        ax=axs[2, 0])

mot_prefrence_groups[5].plot(kind="barh", 
                        title=list(mot_prefrence_groups[5].index[0])[0],
                        color="#E2C43F",
                        edgecolor="black",
                        ax=axs[2, 1])


plt.tight_layout()
```

The visualization above shows that petrol cars are the most preferred mode of transport among festival-goers from all locations. Buses come in second, though with a noticeable gap.

However, a more detailed analysis could provide better insights. I will analyze the preferred modes of transport during peak and non-peak hours. I will use the two subsets of data I had segmented earlier: the peak and no-peak datasets.

```python
# Loop over the list of "location" labels for the "peak_subset"
mot_prefrence_groups_peak = []
for origin in origins_list:
    # Subset "peak_subset" into a set containing only one "origin" location
    origin_mask_peak = peak_subset[peak_subset["Origin"] == origin]
    # Group the different MoTs based on the origin
    origin_groups_peak = origin_mask_peak.groupby("Origin")["Mode of Transport"].value_counts()
    # Visually represent the MoT preference by people from different origin locations
    mot_prefrence_groups_peak.append(origin_groups_peak)
```

```python
# Loop over the list of "location" labels for the "no_peak_subset"
mot_prefrence_groups_no_peak = []
for origin in origins_list:
    # Subset "no_peak_subset" into a set containing only one "origin" location
    origin_mask_no_peak = no_peak_subset[no_peak_subset["Origin"] == origin]
    # Group the different MoTs based on the origin
    origin_groups_no_peak = origin_mask_no_peak.groupby("Origin")["Mode of Transport"].value_counts()
    # Visually represent the MoT preference by people from different origin locations
    mot_prefrence_groups_no_peak.append(origin_groups_no_peak)
```
